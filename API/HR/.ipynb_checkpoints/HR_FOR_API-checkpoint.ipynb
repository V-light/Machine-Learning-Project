```python
y = df["left"]
X = df.drop("left", axis=1)
```

```python
# X["time_spend_company"] = pd.cut(X["time_spend_company"], bins=[0,2,5, 1000], labels=[2,3,1])
# X["time_spend_company"] = X["time_spend_company"].astype('int64')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
# from sklearn.linear_model import LogisticRegression
# model = LogisticRegression()
# from sklearn.tree import DecisionTreeClassifier
# model = DecisionTreeClassifier()
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=300, random_state=1)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_prd = model.predict(X_test)
```

```python
confusion_matrix(y_test, y_prd)
```
