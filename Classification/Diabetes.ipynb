```python
import pandas as pd
df = pd.read_csv("diabetes.csv")
```

```python
df.sample(25)
```

```python
def get_preg(preg):
    var =0
    
    if preg >10:
        var =1
    else :
        var =0
    return var
```

```python
def get_age(age):
    var =0
    age = int(age)
    if age >35:
        var =1
    else :
        var =0
    return var
```

```python
df["Age_title"] = df["Age"].apply(get_age)
df["preg_title"]  = df["Pregnancies"].apply(get_preg)
```

```python
df.info()
```

```python
y = df["Outcome"]
X = df.drop("Outcome", axis = 1)
# from sklearn.preprocessing import StandardScaler
# ss = StandardScaler()
# X = ss.fit_transform(X)
```

```python
# from sklearn.linear_model import LinearRegression
# model = LinearRegression()
# from sklearn.ensemble import RandomForestClassifier
# model = RandomForestClassifier(random_state=1)
from xgboost import XGBClassifier
model = XGBClassifier(eta = 0.3, n_estimators=47, random_state=13, learning_rate=0.1)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.2,random_state = 16)
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import numpy as np

random_grid = {'random_state': [int(x) for x in np.linspace(start=1, stop=20, num = 20)]}
# random_grid = {'eta': [x for x in np.linspace(start=0, stop=0.4, num = 10)]}
# rf_random = RandomizedSearchCV(estimator = model, param_distributions = random_grid, cv = 4)
rf_random = GridSearchCV(model, random_grid, cv=5)
rf_random.fit(X_train, y_train)
print(rf_random.best_score_)
rf_random.best_params_
```
