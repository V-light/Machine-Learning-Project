```python
import pandas as pd
df = pd.read_csv("titanic.csv")
```

```python
df.head(15)
# df = pd.get_dummies(df)
df.info()
df.shape
df.head(15)
```

```python
df.head(20)
df1 =pd.get_dummies(df["Embarked"])
df2 = pd.get_dummies(df["Sex"])
df["Cabin"] = df["Cabin"] =="NaN"
```

```python
df = pd.concat([df, df1,df2], axis = 1)

```

```python
df.head()
df["Age"].mean()
# df["Age"].fillna(df["Age"].mean(), inplace =True)
df["Age"].interpolate(inplace=True)
```

```python
y = df['Survived']
x  = df.drop(["Ticket","Name","Embarked", "Sex", "Survived"], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=11)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=350, random_state=5)
```

```python
model.fit(x_train,y_train )
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_prd = model.predict(x_test)
confusion_matrix(y_test, y_prd)
```
