```python
import pandas as pd
df = pd.read_csv("processed2.csv.")
df.head(20)
```

```python
def get_top(df, thr = 15, no = 10):
    df2 = df.groupby("DISH_NAME").LIKE.mean()
    df4 = pd.DataFrame(df2)
    df4["no"] = df.groupby("DISH_NAME").LIKE.count()
    df5 = df4[df4["no"]>thr]
    maxno = df5["no"].max()
    df5["LIKE"] += df5["no"]/maxno -0.2
    
    return df5.sort_values(["LIKE"], ascending = [0]).head(no)
  
get_top(df, thr = 15, no = 10)
```

```python
food_matrix = df.pivot_table(index = "YOUR_NAME", columns = "DISH_NAME", values = "LIKE")
# food_matrix.head()
```

```python
def get_sim(food_matrix ,dish_name, thr = 20, no = 5 ):
    FM_like = food_matrix[dish_name]
    sim_to_FM_like = food_matrix.corrwith(FM_like)
    corr_sim_FM_food = pd.DataFrame(sim_to_FM_like, columns = ["correlation"])
    corr_sim_FM_food.dropna(inplace =True)
    dftemp = df.groupby("DISH_NAME").LIKE.count()
    dftemp = pd.DataFrame(dftemp)
    corr_sim_FM_food = corr_sim_FM_food.merge(dftemp , on="DISH_NAME")
    
    corr_sim_FM_food =   corr_sim_FM_food[corr_sim_FM_food["LIKE"]>thr]
    return corr_sim_FM_food.sort_values(by="correlation", ascending = False).head(no)

get_sim(food_matrix ,"Chicken Biryani", thr = 20, no = 5 )
```

```python
def get_sim_user_reviews(df, food_matrix, Your_Name, thr =20):
    F_user  = food_matrix.loc[Your_Name]
    sim_to_F = food_matrix.corrwith(F_user, axis = 1)
    sim_to_F = pd.DataFrame(sim_to_F, columns = ["correlation"])
    sim_to_F.dropna(inplace= True)
    coUsers = df.groupby("YOUR_NAME").LIKE.count()
    coUsers = pd.DataFrame(coUsers)
  
    sim_to_F =  sim_to_F.join(coUsers)
    
    sim_to_F     = sim_to_F[ sim_to_F["LIKE"]>thr]
    
    sim_to_F = sim_to_F.sort_values(by="correlation", ascending = False)
    
    
    

    return sim_to_F
get_sim_user_reviews(df, food_matrix ,"Moin",thr =20)
```

```python
import pandas as pd
df = pd.read_csv("processed2.csv")
df.head()
```

```python
def get_top_rated(df , thr= 20, no= 10):
    df2 = df.groupby("DISH_NAME").LIKE.mean()
    df2 = pd.DataFrame(df2)
    df2["NO"] = df.groupby("DISH_NAME").LIKE.count()
    df2 =  df2[ df2["NO"]>thr]
    
    return df2.sort_values(by= "LIKE" ,ascending = False).head(no)

get_top_rated(df, thr = 20, no = 10)
```

```python
def get_top(df, thr = 15, no = 10):
    df2 = df.groupby("DISH_NAME").LIKE.mean()
    df4 = pd.DataFrame(df2)
    df4["no"] = df.groupby("DISH_NAME").LIKE.count()
    df5 = df4[df4["no"]>thr]
    maxno = df5["no"].max()
    df5["LIKE"] += df5["no"]/maxno -0.2
    
    return df5.sort_values(["LIKE"], ascending = [0]).head(no)
  
get_top(df, thr = 15, no = 10)
```

```python
def get_curated(df, thr = 20, no =10):
    df3 = df.groupby("DISH_NAME").LIKE.mean()
    df3 = pd.DataFrame(df3)
    df3["NO"] = df.groupby("DISH_NAME").LIKE.count()
    df3 =   df3[df3["NO"]>thr]
    maxno = df3["NO"].max()
    
    df3["LIKE"]+= df3["NO"]/maxno -0.2
    
    return df3.sort_values(by = "LIKE", ascending  = False).head(no)
get_curated(df)
```

```python
food_metrics = df.pivot_table(index = "YOUR_NAME", columns = "DISH_NAME", values = "LIKE")
```

```python
def get_sim(food_metrics , food_name, thr = 20, no = 10):
    F_Name = food_metrics[food_name]
    Sim_Food = food_metrics.corrwith(F_Name)
    df5 = pd.DataFrame(Sim_Food, columns = ["correlation"])
    temp_df = df.groupby("DISH_NAME").LIKE.count()
    temp_df = pd.DataFrame(temp_df)
    df5 = df5.merge(temp_df, on ="DISH_NAME")
    df5.dropna(inplace =True)
    df5  =   df5[df5["LIKE"]>thr]
    return df5.sort_values(by= "correlation",ascending = False).head(no)
get_sim(food_metrics , "Chicken Biryani", thr = 20, no = 10)
```

```python
def  get_sim_user (df,food_metrics, uid, thr = 20):
    user_f = food_metrics.loc[uid]
    similar_user = food_metrics.corrwith(user_f, axis = 1)
    df6 = pd.DataFrame(similar_user)
    df6.dropna(inplace = True)
    coUsers = df.groupby("YOUR_NAME").LIKE.count()
    coUsers = pd.DataFrame(coUsers )
    coUsers.columns = ["no"]
    df6 = df6.join(coUsers["no"])
    return df6
```

```python
def get_sim_user_top_rated(df, food_metrics, uid, thr=20):
    newDf = get_sim_user(df, food_metrics, uid, thr=20)
    return get_top_rated(newDf, 10, 10)
```
